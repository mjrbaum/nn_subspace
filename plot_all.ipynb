```python
import os
import pandas as pd
from matplotlib import pyplot as plt
```

```python
scnn_dict = {'Small CNN (2,558 params)': {'file': 'fscnn_lr5e-4_epoch40_base.csv'},
              'Sub (250 params)': {'file': 'fscnn_lr5e-4_epoch40_kl250.csv'},
              'Sub WR (250 Params)': {'file': 'fscnn_lr5e-4_epoch40_kl250_rand5.csv'},
              'Sub 10 Epochs -> WR (250 params)': {'file': 'fscnn_lr5e-4_epoch40_kl250_10nrand1.csv'}}

scnn_dir = 'cnn'
```

```python
mcnn_dict = {'Mid CNN (36,656 params)': {'file': 'fmcnn_lr5e-4_epoch40_base.csv'},
              'Sub (6,000 params)': {'file': 'fmcnn_lr5e-4_epoch40_kl6000.csv'},
              'Sub WR (6,000 params)': {'file': 'fmcnn_lr5e-4_epoch40_kl6000_rand5.csv'},
              'Sub 10 Epochs -> WR (6,000 params)': {'file': 'fmcnn_lr5e-4_epoch40_kl6000_10nrand1.csv'}}

mcnn_dir = 'cnn'
```

```python
mmlp_dict = {'Mid MLP (80,730 params)': {'file': 'fmmlp_lr5e-4_epoch40_base_batch16.csv'},
              'Sub (4,000 params)': {'file': 'fmmlp_lr5e-4_epoch40_kl4000.csv'},
              'Sub WR (4,000 params)': {'file': 'fmmlp_lr5e-4_epoch40_kl4000_rand5.csv'},
              'Sub 10 Epochs -> WR (4,000 params)': {'file': 'fmmlp_lr5e-4_epoch40_kl4000_10nrand1.csv'}}

mmlp_dir = 'mlp'
```

```python
train_dict = mmlp_dict
dir = mmlp_dir
```

```python
for key in train_dict.keys():
    data = pd.read_csv(os.path.join(dir, train_dict[key]['file']), index_col=0)
    train_dict[key]['epoch'] = data['epoch']
    train_dict[key]['train_loss'] = data['train_loss']
    train_dict[key]['test_loss'] = data['test_loss']
    train_dict[key]['test_acc'] = data['test_acc']
```

```python
for key in train_dict.keys():
    plt.plot(train_dict[key]['epoch'], train_dict[key]['train_loss'], label=key)
plt.legend()
```

```python
for key in train_dict.keys():
    plt.plot(train_dict[key]['epoch'], train_dict[key]['test_loss'], label=key)
plt.legend()
```

```python
for key in train_dict.keys():
    plt.plot(train_dict[key]['epoch'], train_dict[key]['test_acc'], label=key)
plt.legend()
```

```python
dicts = [scnn_dict, mcnn_dict, mmlp_dict]
dirs = [scnn_dir, mcnn_dir, mmlp_dir]

fig, axs = plt.subplots(3,3, figsize=(16,12), sharex=True)
fig.set_tight_layout(True)

# Set global font sizes
plt.rcParams.update({
    'font.size': 12,          # General text size
    'axes.titlesize': 16,     # Subplot title size
    'axes.labelsize': 16,     # X and Y label size
    'xtick.labelsize': 10,    # X tick label size
    'ytick.labelsize': 10,    # Y tick label size
    'legend.fontsize': 10,    # Legend size
    'figure.titlesize': 16    # Figure suptitle size
})

for i in range(3):
    train_dict = dicts[i]
    dir = dirs[i]
    for key in train_dict.keys():
        data = pd.read_csv(os.path.join(dir, train_dict[key]['file']), index_col=0)
        train_dict[key]['epoch'] = data['epoch']
        train_dict[key]['train_loss'] = data['train_loss']
        train_dict[key]['test_loss'] = data['test_loss']
        train_dict[key]['test_acc'] = data['test_acc']

    for key in train_dict.keys():
        axs[i,0].plot(train_dict[key]['epoch'], train_dict[key]['train_loss'], alpha=0.5, label=key)
    axs[i,0].legend()

    for key in train_dict.keys():
        axs[i,1].plot(train_dict[key]['epoch'], train_dict[key]['test_loss'], alpha=0.5, label=key)
    axs[i,1].legend()

    for key in train_dict.keys():
        axs[i,2].plot(train_dict[key]['epoch'], train_dict[key]['test_acc'], alpha=0.5, label=key)
    axs[i,2].legend(loc='lower right')

axs[0,0].set_title('Training Loss')
axs[0,1].set_title('Validation Loss')
axs[0,2].set_title('Validation Accuracy')

axs[0,0].set_ylabel('Small CNN')
axs[1,0].set_ylabel('Mid-sized CNN')
axs[2,0].set_ylabel('Mid-sized MLP')

axs[2,0].set_xlabel('Epoch')
axs[2,1].set_xlabel('Epoch')
axs[2,2].set_xlabel('Epoch')

for i in range(3):
    for j in range(2):
        axs[i,j].set_ylim(0,2.5)

for i in range(3):
    axs[i,2].set_ylim(0,1)
```
