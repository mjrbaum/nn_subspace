```python
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision.transforms import ToTensor
from torchinfo import summary

import numpy as np

import pandas as pd

from matplotlib import pyplot as plt

from utils.utils import count_parameters, get_params, train, test, run_svd, Subspace_Generator
from utils.nn_arch import SimpleCNN, MidCNN, Subspace_model, Medium_MLP

import random
import os
```

```python
def set_seed(seed=212):
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
    torch.cuda.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)
    
    # Force deterministic behavior
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False
    
    # Optional: ensure all operations are deterministic
    # torch.use_deterministic_algorithms(True)
    
    # Set hash seed for python
    os.environ['PYTHONHASHSEED'] = str(seed)

set_seed(28)
```

```python
# Download training data from open datasets.
training_data = datasets.MNIST(
    root="data",
    train=True,
    transform=ToTensor(),
)

# Download test data from open datasets.
test_data = datasets.MNIST(
    root="data",
    train=False,
    transform=ToTensor(),
)
```

```python
from torch.utils.data import random_split

# Load the full training set
dataset_full = datasets.MNIST(root='data', train=False, transform=ToTensor())

# Define split lengths
train_size = 500
val_size = 9500
training_data, test_data = random_split(dataset_full, [train_size, val_size])
```

```python
batch_size = 16
epochs = 40
lr = 5e-4
pre = 3
rand_sub = 0
M = 6000

# Create data loaders
train_dataloader = DataLoader(training_data, batch_size=batch_size)
test_dataloader = DataLoader(test_data, batch_size=batch_size)

for X, y in test_dataloader:
    print(f"Shape of X [N, C, H, W]: {X.shape}")
    print(f"Shape of y: {y.shape} {y.dtype}")
    break
```

```python
device = torch.accelerator.current_accelerator().type if torch.accelerator.is_available() else "cpu"
print(f"Using {device} device")

#model = SimpleCNN().to(device)
#model = MidCNN().to(device)
model = Medium_MLP().to(device)
print(model)
```

```python
print('Model total params and trainable params: ', count_parameters(model))
X, y = next(iter(train_dataloader))
summary(model, input_size=(batch_size, X.shape[1], X.shape[2], X.shape[3]))
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=lr)
```

```python
loss_train = []
loss_test = []
acc_test = []

for t in range(pre):
    print(f"Epoch {t+1}\n-------------------------------")
    loss_tr, params = train(train_dataloader, model, loss_fn, optimizer, device=device, return_loss=True, return_params=True)
    if t == 0:
        params_all = params
    else:
        params_all = np.concat([params_all, params], axis=1)
    loss_train.append(loss_tr)

    loss_te, acc = test(test_dataloader, model, loss_fn, device=device)
    loss_test.append(loss_te)
    acc_test.append(acc)
print("Done!")
```

```python
print(params_all.shape)
```

```python
vecs, vals = run_svd(params_all, remove_first=100)
```

```python
M = 4000
```

```python
sub_gen = Subspace_Generator(vecs, vals, M)
if rand_sub:
    subspace = sub_gen.rand_subspace()
    #subspace = sub_gen.nonweighted_rand_subspace()
else:
    subspace = sub_gen.max_subspace()
```

```python
model1 = Subspace_model(model=model, vecs=subspace)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model1.params_d, lr=lr)
```

```python
for p in model1.parameters():
    print(p.size())
```

```python
for t in range(epochs-pre):
    print(f"Epoch {t+1}\n-------------------------------")
    if rand_sub and t > 0:
        if t%rand_sub == 0:
            #subspace = sub_gen.rand_subspace()
            subspace = sub_gen.nonweighted_rand_subspace()
            model1.new_subspace(subspace)

    if t >= 10:
        subspace = sub_gen.nonweighted_rand_subspace()
        model1.new_subspace(subspace)

    loss_tr = train(train_dataloader, model1, loss_fn, optimizer, device=device, return_loss=True)
    loss_train.append(loss_tr)
    
    loss_te, acc = test(test_dataloader, model1, loss_fn, device=device)
    loss_test.append(loss_te)
    acc_test.append(acc)
print("Done!")
```

```python
plt.plot(range(1,epochs+1), loss_train, label='Train Loss')
plt.plot(range(1,epochs+1), loss_test, label='Val Loss')
plt.title('Loss vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.ylim(0,2.5)
plt.legend()
```

```python
plt.plot(range(1,epochs+1), acc_test)
plt.title('Accuracy on Validation Set vs. Epoch')
plt.xlabel('Epoch')
plt.ylabel('Classification Accuracy')
plt.ylim(0,1)
```

```python
data_train = pd.DataFrame()

data_train['epoch'] = range(1,epochs+1)
data_train['train_loss'] = loss_train
data_train['test_loss'] = loss_test
data_train['test_acc'] = acc_test

print(data_train.head())

data_train.to_csv('mlp/fmmlp_lr5e-4_epoch40_kl4000_10nrand1.csv')
```
